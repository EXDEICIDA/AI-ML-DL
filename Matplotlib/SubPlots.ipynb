```python
import  numpy as np
import matplotlib.pyplot as plt
```

```python
a  = np.linspace(0,10,11)
```

```python
b  = a**4 # exponential
```

```python
x = np.arange(0,10) # Linear Growth
```

```python
y = 2*x
```

```python
z  = np.linspace(0, 10, 100)
r  = np.sin(z) * np.exp(0.1 * z) #Sine Wave Growth
```

```python
u = np.linspace(-6, 6, 10)
o = 1 / (1 + np.exp(-x))  # S-curve growth logistic
```

## Sub Plots

```python
fig,axes = plt.subplots(nrows=2,ncols=2)

axes[0][0].plot(x,y)
axes[0][0].set_title("Linear Growth")
axes[0][0].set_xlabel("x axis")
axes[0][0].set_ylabel("y axis")
axes[0][1].plot(a,b)
axes[0][1].set_title("Exponential Growth")
axes[0][1].set_xlabel("x axis")
axes[0][1].set_ylabel("y axis")
axes[1][0].plot(z,r)
axes[1][0].set_title("Sine Wave Growth")
axes[1][0].set_xlabel("x axis")
axes[1][0].set_ylabel("y axis")
axes[1][1].plot(u,o)
axes[1][1].set_title("Logistic Growth")
axes[1][1].set_xlabel("x axis")
axes[1][1].set_ylabel("y axis")
plt.tight_layout()
```
