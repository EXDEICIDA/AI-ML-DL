```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,11,10)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x,label = 'X vs X')
ax.plot(x,x**2,label = 'X vs X^2')
ax.plot(x,x**3,label = 'X vs X^3')
ax.plot(x,x**4,label = 'X vs X^3')

ax.legend(loc = 0

          )
plt.show()

```
