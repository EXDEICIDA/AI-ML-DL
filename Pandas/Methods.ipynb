```python
def getQuality(total_bill, tip):
  if tip/total_bill > 0.25:
    return "Generous"
  else:
    return "Other"
```

```python
getQuality(16.99, 1.01)
```

```python
df[['total_bill','tip']].apply(lambda df: getQuality(df['total_bill'], df['tip']), axis=1)
```

```python
df['Quality'] = np.vectorize(getQuality)(df['total_bill'], df['tip'])
df.head()
```
